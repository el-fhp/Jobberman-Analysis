```python
import numpy as np
import pandas as pd
df = pd.read_csv(r"C:\Users\hp\OneDrive\Desktop\Data Mining\Jobberman Job Listings2", index_col = 0)
```

```python
def new_Salary(x):
    if pd.isna(x) or str(x) == 'Confidential':
        return np.nan
    x = str(x).replace('NGN','').replace(',','').strip()
    if '-' in x:
        low, high = x.split('-')
        return (float(low)+ float(high)) / 2
    else:
        return float(x)
```

```python
df['Salary'] = df['Salary'].map(new_Salary)
```

```python
df.columns
```

```python
df.columns = ['Job Title', 'Company_Name', 'Location', 'Employment_Type', 'old_salary',
       'Job_Description', 'Salary']
df.drop('old_salary', axis = 1, inplace = True)
```

```python
df.head(5)
```

```python
# Locations with the highest Job offerings
lg = df.groupby('Location')['Job Title'].count().sort_values(ascending = False)
lg
```

```python
# Highest appearing Job role
df.groupby('Job_Description')['Job Title'].count().sort_values(ascending = False).head(5)
```

```python
# Companies hiring the most
df.groupby('Company_Name')['Job Title'].count().sort_values(ascending = False)
```

```python
# Higest and lowest paying Job role
df.groupby('Job_Description')['Salary'].mean().sort_values(ascending = False)
```

```python
# locations with the higest salary
df.groupby('Location')['Salary'].mean().sort_values(ascending = False)
```
